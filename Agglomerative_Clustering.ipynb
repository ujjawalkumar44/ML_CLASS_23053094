## Agglomerative Clustering Implementation

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

df = pd.DataFrame(X, columns=['Feature 1', 'Feature 2'])

print("Sample Data Head:")
display(df.head())

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Feature 1', y='Feature 2', data=df, palette='viridis', legend=False)
plt.title('Original Sample Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.show()
```

Now, let's apply Agglomerative Clustering to this data.

```python
from sklearn.cluster import AgglomerativeClustering

agg_clustering = AgglomerativeClustering(n_clusters=4)
clusters = agg_clustering.fit_predict(X)

df['Cluster'] = clusters

print("Data with Cluster Labels Head:")
display(df.head())

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Feature 1', y='Feature 2', hue='Cluster', data=df, palette='viridis', legend='full')
plt.title('Agglomerative Clustering Results')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.show()
```
