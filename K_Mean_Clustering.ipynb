### K-Means Clustering Implementation

K-Means clustering is an unsupervised learning algorithm used to partition `n` observations into `k` clusters, where each observation belongs to the cluster with the nearest mean (centroid), serving as a prototype of the cluster. This results in a partitioning of the data space into Voronoi cells.

Here are the general steps we will follow:
1.  **Import Libraries**: Import necessary libraries like `sklearn` for K-Means and `matplotlib` for visualization.
2.  **Generate Sample Data**: Create a synthetic dataset to demonstrate the clustering.
3.  **Apply K-Means**: Initialize and fit the K-Means model to the data.
4.  **Visualize Results**: Plot the clusters and their centroids.

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

print("Libraries imported successfully.")
```

```python
n_samples = 300
n_features = 2
n_clusters = 4

X, y = make_blobs(n_samples=n_samples, n_features=n_features, centers=n_clusters, random_state=42)

print(f"Generated {n_samples} samples with {n_features} features for {n_clusters} clusters.")

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], s=50, cmap='viridis')
plt.title('Generated Sample Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
kmeans.fit(X)

labels = kmeans.labels_
centroids = kmeans.cluster_centers_

print("K-Means clustering completed.")
print("Cluster centroids:\n", centroids)
```

```python
plt.figure(figsize=(10, 8))

for i in range(n_clusters):
    plt.scatter(X[labels == i, 0], X[labels == i, 1], s=100, label=f'Cluster {i+1}')

plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='red', marker='X', label='Centroids')

plt.title('K-Means Clustering Results')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```
